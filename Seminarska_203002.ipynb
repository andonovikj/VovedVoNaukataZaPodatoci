#Correlation Function

```python
def findCorrelation(correlation_type):
  datasets = [dfApple, dfDJIOilAndGas, dfUSDollarIndex, dfBNB, dfMeta]
  datasetNames = ["dfApple", "dfDJIOilAndGas", "dfUSDollarIndex", "dfBNB", "dfMeta"]
  correlationsList = []
  for i in range(0, 5):
    for j in  range(i+1, 5):
      correlationIndex = datasets[i]["Price"].corr(datasets[j]["Price"], method=correlation_type)
      correlationBetweenDatasets = [datasetNames[i], datasetNames[j], correlationIndex]
      correlationsList.append(correlationBetweenDatasets)
  return correlationsList
```

#Pearson Correlation

```python
pearsonCorrelations = pd.DataFrame(findCorrelation("pearson"), columns = ["Dataset 1", "Dataset 2", "Pearson Correlation"])
pearsonCorrelations = pearsonCorrelations.sort_values('Pearson Correlation', ascending=False)
print(pearsonCorrelations)
```

#Kendall Correlation

```python
kendallCorrelations = pd.DataFrame(findCorrelation("kendall"), columns = ["Dataset 1", "Dataset 2", "Kendall Correlation"])
kendallCorrelations = kendallCorrelations.sort_values('Kendall Correlation', ascending=False)
print(kendallCorrelations)
```

#Spearman Correlation

```python
spearmanCorrelations = pd.DataFrame(findCorrelation("spearman"), columns = ["Dataset 1", "Dataset 2", "Spearman Correlation"])
spearmanCorrelations = spearmanCorrelations.sort_values('Spearman Correlation', ascending=False)
print(spearmanCorrelations)
```

#Корелација со зададен window

```python
#calculate the moving averages for 7 days: capture the average data change in statistics over time
#select the Price column, for the rolling method, the window is 7 and chain with the mean
dfDJIOilAndGas["7D"] = dfDJIOilAndGas["Price"].rolling(window = 7).mean()
dfDJIOilAndGas.head(10)
```

```python
windowCorrelation = dfDJIOilAndGas["Price"].corr(dfUSDollarIndex["7D"], method='pearson')
windowCorrelation
```
